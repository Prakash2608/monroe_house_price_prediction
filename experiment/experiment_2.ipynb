```python
def feature_engineering_price_per_sqft(df):

    df['price_per_sqft'] = df['price']*100000/df['total_sqft']
    
    return df
            
        
```

```python
def feature_engineering_location(df):

    df['location'] = df['location'].apply(lambda x: x.strip())
    location_stats = df['location'].value_counts(ascending=False)
    location_stats_less_than_10 = location_stats[location_stats<=10]
    df.location = df.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
    
    return df
            
       
```

```python
def outlier_price_per_sqft(df):
       
    df = df[~(df.total_sqft/df.bhk<300)]
    
    df_out = pd.DataFrame()
    
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    
    return df
            
       
```

```python
def outlier_bhk(df):
        
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
                
    return df.drop(exclude_indices,axis='index')
```

```python
def outlier_bathroom(df):
            
    df = df[df.bath<df.bhk+2]
    df = df.drop(['size','price_per_sqft'],axis='columns')
    
    return df  
```

```python
def location_one_hot_encoding(df):
        
    location_dummies = pd.get_dummies(df['location'])
    df = pd.concat([df, location_dummies.drop('other', axis = 'columns')], axis='columns')
    df = df.drop('location',axis='columns')
    
    return df
```

```python
def get_data_transform(df):
    df = drop_unwanted_columns(df)
    df = drop_null_values(df)
    df = feature_engineering_bhk(df)
    df = feature_engineering_total_sqft(df)
    df = feature_engineering_price_per_sqft(df)
    df = feature_engineering_location(df)
    df = outlier_price_per_sqft(df)
    df = outlier_bhk(df)
    df = outlier_bathroom(df)
    df = location_one_hot_encoding(df)
    
    return df
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
import pandas as pd
```

```python
class DropUnwantedColumns(BaseEstimator, TransformerMixin):
    def __init__(self, columns_to_drop):
        self.columns_to_drop = columns_to_drop

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X.drop(columns=self.columns_to_drop)
```

```python
class DropNullValues(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X.dropna()
```

```python
class FeatureEngineeringBHK(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X['bhk'] = X['size'].apply(lambda x: int(x.split(' ')[0]))
        return X
```

```python
class ConvertSqftToNum(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        def convert_sqft_to_num(x):
            tokens = x.split('-')
            if len(tokens) == 2:
                return (float(tokens[0]) + float(tokens[1])) / 2
            try:
                return float(x)
            except:
                return None
        X['total_sqft'] = X['total_sqft'].apply(convert_sqft_to_num)
        X = X[X['total_sqft'].notnull()]
        return X
```

```python
class FeatureEngineeringPricePerSqft(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X['price_per_sqft'] = X['price'] * 100000 / X['total_sqft']
        return X
```

```python
class FeatureEngineeringLocation(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X['location'] = X['location'].apply(lambda x: x.strip())
        location_stats = X['location'].value_counts(ascending=False)
        location_stats_less_than_10 = location_stats[location_stats <= 10]
        X['location'] = X['location'].apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
        return X
```

```python
def remove_spaces(input_df):
    input_df['Totalcharges'] = input_df['Total_charges'].replace(r'^\s*$', np.nan, regex= True)
    return input_df
```
